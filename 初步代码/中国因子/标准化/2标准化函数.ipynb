# 进行标准化处理
相同的行业中，所有的因子进行标准化。
除了[MARKETTYPE  TRDMNT_y  INDCDZX TRDMNT STKCD    date]

```python
df_standard.dtypes
```

```python
df = pd.read_csv('hangyedata.csv', encoding='utf-8')
df_standard=df
```

```python
import numpy as np
import pandas as pd

# 1️⃣ 明确哪些列不能动
exclude_cols = [
    'MARKETTYPE', 'TRDMNT_y', 'INDCDZX',
    'TRDMNT', 'STKCD', 'date'
]

# 2️⃣ 自动识别「因子列」
factor_cols = [
    c for c in df_standard.columns
    if c not in exclude_cols
    and pd.api.types.is_numeric_dtype(df_standard[c])
]

print("将被标准化的因子列数量:", len(factor_cols))

# 3️⃣ 行业 × 月份 内做 z-score
df_standard[factor_cols] = (
    df_standard
    .groupby(['INDCDZX', 'TRDMNT'])[factor_cols]
    .transform(lambda x: (x - x.mean()) / x.std(ddof=0))
)
print("标准化完成！")
print(df_standard[factor_cols].head())

```
